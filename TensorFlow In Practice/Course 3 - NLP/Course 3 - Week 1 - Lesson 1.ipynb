```
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
sentences=[
    'I love my oneplus',
    'I love my cat',
    'You love my cat!!!',
    'Do you think my oneplus  is amazing ?'
]
tokenizer=Tokenizer(num_words=100,oov_token="<OOV>")
tokenizer.fit_on_texts(sentences)
word_index=tokenizer.word_index
sequences=tokenizer.texts_to_sequences(sentences)
padded=pad_sequences(sequences)
print(word_index)
print(sequences)
print(padded)
padded=pad_sequences(sequences,padding='post')
print('padding=post means zeros after the main data part');
print(word_index)
print(sequences)
print(padded)
padded=pad_sequences(sequences,padding='post',truncating='post')
print('padding+Truncating');
print(word_index)
print(sequences)
print(padded)
```

```
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
sentences=[
    'I love my oneplus',
    'I love my cat',
    'You love my cat!!!',
    'Do you think my oneplus  is amazing ?'
]
tokenizer=Tokenizer(num_words=100,oov_token="<OOV>")
tokenizer.fit_on_texts(sentences)
word_index=tokenizer.word_index
sequences=tokenizer.texts_to_sequences(sentences)
padded=pad_sequences(sequences)
padded=pad_sequences(sequences,padding='post',maxlen=2,truncating='post')
print('padding+Truncating+maxlen=2 so that the padding cut all others into similar 2 length matrix');
print(word_index)
print(sequences)
print(padded)
```
